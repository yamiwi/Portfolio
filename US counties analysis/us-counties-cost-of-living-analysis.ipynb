## Get familiarized with dataset

| Variable | Description |
| --- | --- |
| case_id | Unique identifier for each areaname. |
| state | The state where the county is located.

| total_cost | Total estimated annual cost of living for the family type in the county. 
| median_family_income | Median annual income for families in the county.

## Split family member count column into 2 separate columns of number of parents and number of children

```python
# Make a copy of the dataset
df1 = df[~df['median_family_income'].isna()].copy()

# Split family member count into 2 separate columns by 'p' delimiter
df1[['number_of_parents', 'number_of_children']] = df1['family_member_count'].str.split('p', expand=True)

# Get rid of tailing 'c' symbol in number of children column
df1['number_of_children'] = df1['number_of_children'].str.strip('c')

# Change data types of these 2 columns into integer type
df1 = df1.astype({'number_of_parents' : np.int8, 'number_of_children': np.int8})

# Make a new column family size 
df1['family_size'] = df1['number_of_parents'] + df1['number_of_children']

# Display dataset
df1.head(10)
```

# Exploratory data analysis

## Explore family income distributions

```python
# Create figure and axes
fig, ax = plt.subplots(1, 2, figsize=(24, 6), gridspec_kw={'width_ratios': [1, 1]})

# Create a histplot of median family income distribution
sns.histplot(data=df1, x='median_family_income', kde=True, bins=50, ax=ax[0])

# Configure the plot
ax[0].set_title('Median family income distribution')
ax[0].set_xlabel('Median family income')
ax[0].set_ylabel('Frequency')

# Create a histplot of median income distribution by metro availability
sns.histplot(data=df1, x='median_family_income', kde=True, bins=50, ax=ax[1], hue='isMetro')

# Configure the plot
ax[1].set_yticks(ticks=np.arange(0, 3001, 500))
ax[1].set_title('Median family income distribution by metro availability')
ax[1].set_xlabel('Median family income')
ax[1].set_ylabel('Frequency')

# Display plots
plt.show()
```

Median family income higher in counties where metro is available in comparsion where metro is not available.

## Explore average family income and average total cost of accomodation by metro evailability

```python
# Group by metro availability and calculate the mean annual living cost for both cases
metro = df1.groupby('isMetro')['total_cost'].agg('mean')
metro_income = df1.groupby('isMetro')['median_family_income'].agg('mean')

# Create figure and axes
fig, ax = plt.subplots(1, 2, figsize=(22, 4))

# Create a barplot
sns.barplot(x=metro_income.index, y=metro_income.values, ax=ax[0])

# Configure the plot
ax[0].set_title('Average annual family income by metro availability')
ax[0].set_xlabel('Is metro')
ax[0].set_ylabel('Average annual family income ')


# Create a barplot 
sns.barplot(x=metro.index, y=metro.values, ax=ax[1])

# Configure the plot
ax[1].set_title('Average total cost by metro availability')
ax[1].set_xlabel('Is metro')
ax[1].set_ylabel('Total cost')

# Display the plot
plt.show()
```

Annual total cost is also higher in counties where metro is available 

## Investigate the amount of records of each state in the dataset

```python
# Set figure size
plt.figure(figsize=(22, 6))

# Create a plot
sns.countplot(data=df1, x='state')

# Configure the plot
plt.title('Dataset records distribution by states')
plt.xlabel('State')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

## Explore average family income by states

```python
# Since there are median family income for every county, we can calculate average family income for every state
median_income_by_states = df1.groupby('state')['median_family_income'].agg('mean')
median_income_by_states = median_income_by_states.map(int)

# Create a figure
plt.figure(figsize=(22, 8))

# Create a plot
ax = sns.barplot(x=median_income_by_states.index, y=median_income_by_states.values)

# Set bar annotations
ax.bar_label(ax.containers[0], fontsize=7)

# Configure the plot
plt.title('Average annual family income by states')
plt.xlabel('State')
plt.ylabel('Annual family income')

# Display the plot
plt.show()
```

## Explore annual total cost of accomodation by states

```python
# Set figure size
plt.figure(figsize=(22, 6))

# Create a plot
sns.boxplot(data=df1, y='total_cost', x='state', orient='v')

# Configure the plot
plt.title('Annual total cost by states')
plt.xlabel('State')
plt.ylabel('Total cost of living')

# Display the plot
plt.show()
```

## Investigate the components of the annual cost of accomodation

```python
# Create figure and axes
fig, ax = plt.subplots(7, 1, figsize=(24, 42))

# Make a temporary index variable
index = 0

# Iterate through dataframe columns
for col in df1.iloc[:, 6:13].columns:
    # Make a plot on every iteration  
    sns.boxplot(data=df1, x='state', y=col, ax=ax[index], orient='v')
    ax[index].set_title('Annual {} by state'.format(' '.join(col.split('_'))))
    ax[index].set_ylabel('{}'.format(' '.join(col.split('_')).capitalize()))
    ax[index].set_xlabel('State')
    index += 1

# Display plots
plt.show()
```

Average annual living costs are slightly higher in counties where metro is present in comparison with counties without metropoliten.

## Explore how total cost changes with family size change

```python
# Create figure and axes
fig, ax = plt.subplots(2, 2, figsize=(19, 10))

# Create a plot
sns.boxplot(data=df1, x='total_cost', y='family_member_count', orient='h', ax=ax[0, 0])

# Configure the plot
ax[0, 0].set_title('Annual total cost by number of parents x number of children')
ax[0, 0].set_xlabel('Total cost')
ax[0, 0].set_ylabel('Number of parents x number of children')

# Create a plot
sns.boxplot(data=df1, x='total_cost', y='number_of_parents', orient='h', ax=ax[0, 1])

# Configure the plot
ax[0, 1].set_title('Annual total cost by number of parents')
ax[0, 1].set_xlabel('Total cost')
ax[0, 1].set_ylabel('Number of parents')


# Create a plot
sns.boxplot(data=df1, x='total_cost', y='family_size', orient='h', ax=ax[1, 0])

# Configure the plot
ax[1, 0].set_title('Annual total cost by family size')
ax[1, 0].set_xlabel('Total cost')
ax[1, 0].set_ylabel('Number of family members')


# Create a plot
sns.boxplot(data=df1, x='total_cost', y='number_of_children', orient='h', ax=ax[1, 1])

# Configure the plot
ax[1, 1].set_title('Annual total cost by number of children')
ax[1, 1].set_xlabel('Total cost')
ax[1, 1].set_ylabel('Number of children')

# Display plots
plt.show()
```
